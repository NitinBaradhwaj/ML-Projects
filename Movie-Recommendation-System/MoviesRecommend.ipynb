# Movie Recommedation System

## What is a Recommedation System
A recommender system, or a recommendation system, is a subclass of information filtering system that seeks to predict the “rating” or “preference” a user would give to an item. They are primarily used in commercial applications.

## Types of Recommendation System

1. Content based: Content-based filtering methods are based on featuralization of items (as oppose to users) and a profile of a user’s utility. 


2. Collaborative Filtering: Collaborative filtering is best suited to problem with known data on users (age, gender, occupation, etc), but lack of data for items or difficult to do feature extraction for items of interest.


3. Hybrid: Given the limitation of both content-based and collaborative methods, and they both address some limitations from the other, it is natural to consider combining these two methods together, which leads to the Hybrid method

We need to change the format to '['Action','Adventure','Fantasy','Science Fiction']'

```python
import ast
def convert(objects):
    ABC=[]
    for i in ast.literal_eval(objects):
        ABC.append(i['name'])
    return ABC

        
```

```python
moviefile['keywords'] = moviefile['keywords'].apply(convert)
```

```python
def convert1(objects):
    ABC=[]
    cotr=0
    for i in ast.literal_eval(objects):
        if cotr!=3:
            ABC.append(i['name'])
            cotr+=1
        else:
            break
    return ABC
```

Now the data is in proper format

Now remove the space between words like

'Sam Worthington' --> 'SamWorthington'

If we keep it as Sam Worthington it'll be two different entities Sam and Worthington 
If we are searching for Sam Harry. Our Model will get confused that's the reasone we remove space between words

```python
def collapse(L):
    L1 = []
    for i in L:
        L1.append(i.replace(" ",""))
    return L1
```

```python
#Apply this for Genre,Cast,Crew,Keywords
```

```python
moviefile['cast'] = moviefile['cast'].apply(collapse)
moviefile['crew'] = moviefile['crew'].apply(collapse)
moviefile['genres'] = moviefile['genres'].apply(collapse)
moviefile['keywords'] = moviefile['keywords'].apply(collapse)

#or you can use this method
#moviefile['genres'] = moviefile['genres'].apply(lambda a:[b.replace(" ","")for b in a])
#moviefile['keywords'] = moviefile['keywords'].apply(lambda a:[b.replace(" ","")for b in a])
#moviefile['cast'] = moviefile['cast'].apply(lambda a:[b.replace(" ","")for b in a])
#moviefile['crew'] = moviefile['crew'].apply(lambda a:[b.replace(" ","")for b in a])
```

```python
import nltk
from nltk .stem.porter import PorterStemmer
pos = PorterStemmer()
```

```python
def stem(texts):
    x=[]
    for a in texts.split():
        x.append(pos.stem(a))
    return " ".join(x)
```

```python
new_data_frame['tags'] = new_data_frame['tags'].apply(stem)
```

```python
from sklearn.feature_extraction.text import CountVectorizer
cv = CountVectorizer(max_features=5000,stop_words='english')
from sklearn.metrics.pairwise import cosine_similarity
```

```python
vectors = cv.fit_transform(new_data_frame['tags']).toarray()
```

```python
def recommend(movie):
    index = new_data_frame[new_data_frame['title'] == movie].index[0]
    distances = sorted(list(enumerate(similar[index])),reverse=True,key = lambda x: x[1])
    for i in distances[1:6]:
        print(new_data_frame.iloc[i[0]].title)
```
